```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Loading Data using Pandas

```python
#Reading exam data 
df1 = pd.read_csv("./data/exam.csv")
df1
```

```python
#Reading assignment data
df2 = pd.read_csv("./data/assignment.csv")
df2
```

```python
#Reading Project data
df3 = pd.read_csv("./data/project.csv")
df3
```

# Data Merged from Exams, Assignments and Project works

```python
final_data = pd.merge(df1, df2)
final_data = pd.merge(final_data, df3)
final_data
```

# After merging all the datas into one sheet, we now calculated the average score on each dataset

```python
final_data['Exam'] = round((final_data['Exam1'] + final_data['Exam2'] + final_data['Exam3'])/3,2)
final_data['Assignment'] = round((final_data['Assignment1'] + final_data['Assignment2'] + final_data['Assignment3'])/3, 2)
final_data['Project'] = round((final_data['Project1'] + final_data['Project2'] + final_data['Project3'] + final_data['Project4'])/4,2)
final_data
```

## Extracting average score of exams, assignments and projects score only from the above final dataframe

```python
grade_final = final_data[['Name','Department','Assignment','Exam','Project']]
grade_final
```

# Calculating Final Grade of the Students and sorting their Letter Grades

```python
def letter_grading(col):
    if col < 60:
        return 'D'
    elif col < 70:
        return 'C'
    elif col < 80:
        return 'B'
    elif col < 90:
        return 'A'
    else:
        return 'A+'
```

```python
grade_final['Final Grade'] = round(grade_final['Exam']*0.6 + grade_final['Assignment']*0.2 + grade_final['Project']*0.2,2)
grade_final['Final Letter Grade'] = grade_final['Final Grade'].apply(letter_grading)
grade_final
```

# Saving the new final sheet after calculating final grade into .csv file as 'gradesheet_final.csv'

```python
grade_final.to_csv('gradesheet_final.csv', index = False)
```

```python
final = pd.read_csv('gradesheet_final.csv')
final
```

# Now It is the time for plotting the reports. 

```python
# final['Final Letter Grade'].value_counts().plot.bar()

def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])
        
plt.figure(figsize=(10, 5))
grade = final['Final Letter Grade'].value_counts(sort=False)
plt.title("Distribution of Letter Grade Among Students")
plt.xlabel('Grades')
plt.ylabel('Frequency of Grades')
plt.bar(grade.index, grade.values)
addlabels(grade.index, grade.values)
```

```python
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])
        
def sorted_and_plot(column):
    plt.figure(figsize=(25, 10))
    df_final = final.groupby('Name').sum()
    sorted_orders = df_final[column]
    sorted_orders = sorted_orders.sort_values(ascending=False)
    plt.bar(sorted_orders.index, sorted_orders.values)
    plt.xticks(rotation = 75)
    plt.ylim(sorted_orders.values.min()-5, sorted_orders.values.max()+5)
    addlabels(sorted_orders.index, sorted_orders.values)
    plt.title('Distribution of Final Grade among Students')
    plt.xlabel('Name of Students')
    plt.ylabel('Grade of the Students')
sorted_and_plot('Final Grade')
```
